# **Variable Names**
A variable name can have a short name or a descripetive name (age, name, total_volume).

**Rules for R variable names:**

- A variable's name has to begin with a letter and can be a combination of letters, numbers, dots (.), or underscores (_). If it starts with a dot (.), it can't have a number right after it.
- A variable name cannot start with a number or underscore (_)
- Variable names are case-sensitive (time, Time and TIME are three distinct variables)
- You can't use certain words like TRUE, FALSE, NULL, or if as names for your variables.

**Note:** Variable names are case-sensitive.




```
# Acceptable variable names:
myname <- "Moni"
my_name <- "Moni"
myName <- "Moni"
MYNAME <- "Moni"
myname2 <- "Moni"
.myname <- "Moni"

# Inappropriate variable names:
2myname <- "Moni"
my-name <- "Moni"
my name <- "Moni"
_my_name <- "Moni"
my_n@ame <- "Moni"
TRUE <- "Moni"
```

In any programming, data type is an important concept.

- A variable can store data of different types.
- In R, you don't have to specify a specific data type when declaring variables, and they can switch types after their initial assignment.


```
var <- 23 # var is of the numeric type
var <- "Apple" # The variable "var" has been changed to the character type, which is also known as a string.
```

**Basic Data Types:**

The fundamental data types in R can be categorized as follows:
- Numeric (3, 5, 2.0, 5.5)
- Integer - (3L, 6L, 10012L, Here the letter "L" declares this as an integer)
- Complex - (2 + 5i, where "i" is the imaginary part)
- Character (a.k.a. string) - ("m", "C is exciting", "True", "12")
- Logical (a.k.a. boolean) - (TRUE or FALSE)

**Note:**  Here in R, you can use **class()** function to know the datatype of the variable.

**Input:**

```
a <- 23.0
class(a)
b <- "Monica"
class(b)
c <- 2+5i
class(c)
d <- TRUE
class(d)
e <- 2L
class(e)
```
